```python
import torch.nn.functional as F
import torch
import torch.utils.data as Data

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# create some data

n = 800
p = 2

# x data (tensor), shape=(100, 1)
x = torch.rand(n*p).reshape(n, p)

# noisy y data (tensor), shape=(100, 1)
y = (x[:,0].pow(2)+x[:,1].pow(4) + 0.2*torch.rand(x[:,0].size())).reshape(n, 1)

print(x.shape)
print(y.shape)
```

```python
# Data paramaters
seq_len = 1000
n_batch = 100
espilon = 0.05 # step_size 

# Create data
x = torch.tensor(np.zeros((seq_len, n_batch, 1))).float() # 1 = number of features
y = torch.tensor(np.zeros((n_batch))).float()

# feed 
for b in range(n_batch):
    for t in range(seq_len):
        x[t, b, 0] = np.sin(((t + b) * 2 * np.pi) / (n_batch/2))
    y[b] = np.sin(((seq_len + b) * 2 * np.pi) / (n_batch/2))
  
```

```python
# Visualise batch
batch_id = 25

plt.plot(x[:, batch_id, 0].numpy(), 'o', label="Features X_t (0 < t < {})".format(seq_len))
plt.plot(seq_len , y[batch_id].numpy(), 'ro', label="Prediction Y_t")
plt.legend()
plt.show()
```

```python
# the algorithm as a class
class VanillaGRU(torch.nn.Module):
    def __init__(self, n_feature, n_hidden):
        super(VanillaGRU, self).__init__()
        self.hidden1 = torch.nn.GRU(n_feature, n_hidden, num_layers=3)
        self.hidden2 = torch.nn.Linear(n_hidden, 1)
        self.predict = torch.nn.Tanh()
        #self.scale = torch.nn.Linear(1, 1, bias=False)
    def forward(self, x):
        output, _ = self.hidden1(x)
        x = self.hidden2(output[-1])
        x = self.predict(x)
        #x = self.scale(x)
        return(x)

# instantiate GRU
neuralnetwork = VanillaGRU(n_feature=1, n_hidden=32)
print(neuralnetwork)

# optimizer
optimizer = torch.optim.SGD(neuralnetwork.parameters(), lr=0.1)

# loss function
lossfunction = torch.nn.MSELoss(reduce = True, size_average = True)
```

```python
for param in neuralnetwork.parameters():
    print(type(param.data), param.size())
```

```python
e = 0
steplist = []
losslist = []
for epoch in range(10):
  
    # predict the result for the input batch
    prediction = neuralnetwork(x)
    # compute the loss between the prediction and the reality
    loss = lossfunction(prediction, y)
    # clear the gradients for next train
    optimizer.zero_grad()
    # backpropagation to compute the gradients
    loss.backward()
    # apply the gradients
    optimizer.step()

    steplist.append(e)
    losslist.append(lossfunction(neuralnetwork(x), y).item())
    e = e + 1
    p

print(losslist)
plt.plot(losslist)
```

```python
step_radians = 0.01
steps_of_history = 200
steps_in_future = 1
index = 0

x = np.sin(np.arange(0, 20*np.pi, step_radians))

seq = []
next_val = []

for i in range(0, len(x) - steps_of_history, steps_in_future):
    seq.append(x[i: i + steps_of_history])
    next_val.append(x[i + steps_of_history])

seq = np.reshape(seq, [-1, steps_of_history, 1])
next_val = np.reshape(next_val, [-1, 1])
print(np.shape(seq))

trainX = np.array(seq)
trainY = np.array(next_val)
```

```python
print(trainX.shape)
trainX = trainX.swapaxes(0,1)
print("X : ", trainX.shape)
print("Y : ", trainY.shape)
```

```python
trainX = torch.Tensor(trainX)
trainY = torch.Tensor(trainY)
print("X :", trainX.shape)
print("Y :", trainY.shape)
```

```python
e = 0
steplist = []
losslist = []
for epoch in range(10):
    # predict the result for the input batch
    prediction = neuralnetwork(trainX)
    # compute the loss between the prediction and the reality
    loss = lossfunction(prediction, trainY)
    # clear the gradients for next train
    optimizer.zero_grad()
    # backpropagation to compute the gradients
    loss.backward()
    # apply the gradients
    optimizer.step()

    steplist.append(e)
    losslist.append(lossfunction(neuralnetwork(trainX), trainY).item())
    e = e + 1
    print("\r loss : {:.2f}".format(losslist[-1]), end = "")
    
print(losslist)
plt.plot(losslist)
```
